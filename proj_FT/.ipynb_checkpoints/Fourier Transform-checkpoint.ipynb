# Discrete Forier Transform

```python
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#takes in numpy array, x (N)x1, which holds data in position/time space, performs naive fourier transform 
#and returns data in frequency space, as well as corresponding frequncy bins, only considering positive
#note can be easily changed to inverse transform by switching sign of the exponential

#done via encoding discrete fourier transform as a matrix ((N)x(N)), the columns representing evaluation of the exp 
#at different indicies n of the data in time/pos space, the rows being the exp evaluated at different indicies k 
#of the data transformed into frequency space


def discrete_FT(x): 
    N = x.shape[0]
    n = np.arange(N)                    #time/pos space indicies (row vect)
    k = n.reshape((N,1))                #frequency space indicies (col vect) 
    M = np.exp((-2j * np.pi * k* n) / N)   #creation of transform matrix
    freq = np.fft.fftfreq(N, d=1/1024)
    return freq[:N//2], np.abs(np.dot(M,x))[:N//2] #application of matrix to data, only want posive for this example
```

```python
def generate_signal(num_samples=1024, sampling_rate=1024, frequencies=[5, 15, 40], amplitudes=[.75, 0.5, 0.3]):
    t = np.linspace(0, num_samples / sampling_rate, num_samples, endpoint=False)  # Time axis
    signal = np.zeros(num_samples)

    # Sum sinusoidal components with given frequencies and amplitudes
    for amplitude, frequency in zip(amplitudes, frequencies):
        signal += amplitude * np.sin(2 * np.pi * frequency * t)
    signal += np.cos(2 * np.pi * 2 * t)
    return t, signal

def plot(title, x, y, x_lab, y_lab, xlim = None): 
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel(x_lab)
    plt.ylabel(y_lab)
    if xlim != None: 
        plt.xlim(0,50)
    plt.show()
```

## Example on Sample Signal

```python
t_steps, sample_signal = generate_signal()
plot('Sample Signal',t_steps, sample_signal, 'time(s)', 'amplitude(arbitray)' )

freq, trans_signal = discrete_FT(sample_signal)
plot('Sample Signal - Frequency Decomposition',freq, trans_signal, 'frequency(Hz)', 'magnitude(arbitray)', xlim = True )
```

# Discrete Fast Fourier Transform - dim = 1

```python
def IFFT(x):
    N = x.shape[0]

    if np.log2(N) % 1 > 0:
        raise ValueError("must be a power of 2")

    # N_min is number of datapoints (in freq domain) normal discrete IFT will be performed on
    N_min = min(N, 32)

    # performing same as above algo, but inverse, and forcing to be broken into blocks of size N_min (length of sub IFT)
    n = np.arange(N_min)     
    k = n[:, None]
    M = np.exp(2j * np.pi * n * k / N_min)  # Positive sign for IFT
    
    # forces correct dimension of x so all data can be multiplied in parallel with desired block size - computing 
    # all sub IFTs at once via matrix multiplication 
    X = np.dot(M, x.reshape((N_min, -1)))   

    # Continually using sub IFTs, via splitting into odd and even parts and calculating 'twiddle factor', then concat
    # them until eniter transform is complete
    while X.shape[0] < N:
        X_even = X[:, :X.shape[1] // 2]
        X_odd = X[:, X.shape[1] // 2:]
        factor = np.exp(1j * np.pi * np.arange(X.shape[0]) / X.shape[0])[:, None]  # Positive sign (twiddle factor)
        X = np.vstack([X_even + factor * X_odd,
                       X_even - factor * X_odd])

    return X.ravel() # flatten then normalize by N 


def FFT(x):
    N = x.shape[0]

    if np.log2(N) % 1 > 0:
        raise ValueError("must be a power of 2")

    # N_min is number of datapoints (in freq domain) normal discrete IFT will be performed on
    N_min = min(N, 32)

    # performing same as above algo, but inverse, and forcing to be broken into blocks of size N_min (length of sub IFT)
    n = np.arange(N_min)     
    k = n[:, None]
    M = np.exp(-2j * np.pi * n * k / N_min)  
    
    # forces correct dimension of x so all data can be multiplied in parallel with desired block size - computing 
    # all sub IFTs at once via matrix multiplication 
    X = np.dot(M, x.reshape((N_min, -1)))   

    # Continually using sub IFTs, via splitting into odd and even parts and calculating 'twiddle factor', then concat
    # them until eniter transform is complete
    while X.shape[0] < N:
        X_even = X[:, :X.shape[1] // 2]
        X_odd = X[:, X.shape[1] // 2:]
        factor = np.exp(-1j * np.pi * np.arange(X.shape[0]) / X.shape[0])[:, None]  #(twiddle factor)
        X = np.vstack([X_even + factor * X_odd,
                       X_even - factor * X_odd])
        
    return X.ravel() 
```

## Example on Sample Signal 

```python
plot('Sample Signal',t_steps, sample_signal, 'time(s)', 'amplitude(arbitray)' )

FFT_trans_signal = FFT(sample_signal)
N = sample_signal.shape[0]
FFT_freq = np.fft.fftfreq(N, d=1/1024)
plot('Sample Signal - Frequency Decomposition',FFT_freq[:N//2], np.abs(FFT_trans_signal)[:N//2], 'frequency(Hz)', 'magnitude(arbitray)', xlim = True )
```

# Extension to dim = 2

```python
#simply just going to apply to both dimensions of input 
def FFT_2D(x, forward = True):
    rows, cols = x.shape
    if (np.log2(rows) % 1 > 0) or (np.log2(cols) % 1 > 0):
        raise ValueError("Both dimensions of the input array must be powers of 2")

    if forward == True: 
        X_rows = np.array([FFT(row) for row in x])    #row basis 
        X_2D = np.array([FFT(col) for col in X_rows.T]).T  #col basis, of the already transformed
    if forward == False: 
        X_rows = np.array([IFFT(row) for row in x])
        X_2D = np.array([IFFT(col) for col in X_rows.T]).T  
        
    return X_2D / (rows * cols) if not forward else X_2D  # Normalize by total number of elements for 2D IFFT

def pow_2_reshape(image_array): 
    img_x, img_y = image_array.shape
    pows = np.arange(0,12)
    pows_of_2 = 2**pows
    
    diff_x = np.abs(pows_of_2 - img_x) 
    diff_y = np.abs(pows_of_2 - img_y)
    closest_x = pows_of_2[np.argmin(diff_x)]
    closest_y = pows_of_2[np.argmin(diff_y)]    
    
    return (closest_x, closest_y)
```

```python
image = Image.open("brain_NMR_image.jpeg").convert("L")  # "L" mode is for grayscale
image_resized = image.resize(pow_2_reshape(np.array(image)))
image_array = np.array(image_resized)

transformed_image = FFT_2D(image_array)
recovered_image = FFT_2D(transformed_image, forward = False)

fig, axs = plt.subplots(2, 2, figsize=(12, 12))
#original image
axs[0, 0].imshow(image_array, cmap="gray")
axs[0, 0].set_title("Original Image")
#image after FFT 
axs[0, 1].imshow(np.log(np.abs(transformed_image) + 1), cmap="gray")
axs[0, 1].set_title("Transformed Image")
#image recovered from aforementioned transformation
axs[1, 0].imshow(np.abs(recovered_image), cmap = 'gray')
axs[1, 0].set_title("Image Recovered (Inverse Transform)")
axs[1, 1].remove() 

plt.tight_layout()
plt.show()
```
